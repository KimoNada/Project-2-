```python
# initializing library with those books
class Library:
    
    
    def __init__(self,listofbooks=[]):
        # set up the library with the books list
        self.listofbooks = listofbooks

    def displayAvailablebooks(self):
        print("The books we have in our library are as follows:")
        print("================================")
        # write a code to print all books
        print(self.listofbooks)
                    
                    
    def lendBook(self,requestedBook):
        # write condition to see if the book is available or not and print the output
        self.requestedBook = requestedBook
        if requestedBook in self.listofbooks :
            print('Requested Book Available', requestedBook)
            self.listofbooks.remove(requestedBook)
        else:
            print('Requested Book Not Available')
            
    def addBook(self,returnedBook):
        # write code to add the book back into your library
        self.returnedBook = returnedBook
        if returnedBook not in self.listofbooks :
            self.listofbooks.append( returnedBook )
            print('Book is returned: ', returnedBook)
        else:
            print('Book is already found')

library=Library(["The Last Battle","The Screwtape letters","The Great Divorce"])

done=False

while done == False:# set condition for the loop :
      print(""" ======LIBRARY MENU=======
            Enter the number of your choice:
            
            1. Display all available books
            2. Lend a book
            3. Return a book
            4. Exit
            """)

      choice=int( input() )

      if choice==1:
            # fill this as you see suitable 
            library.displayAvailablebooks()

      elif choice==2:
            # take the input from the user and use the method you see suitable
            library.lendBook(requestedBook=input("You Asking for: "))

      elif choice==3:
            # take the input from the user and use the method you see suitable
            library.addBook(returnedBook=input("You Returned: "))
            
      else:
            # set a condition to close the loop
            print('Exiting...')
            break
```
